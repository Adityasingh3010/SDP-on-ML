Question 4.8

Negative

```python
max_intensity = 255
(h, w) = img_gs.shape
output = img_gs.copy()
for i in np.arange(h):
  for j in np.arange(w):
    a = output.item(i,j)
    b = max_intensity - a
    output.itemset((i,j), b)
cv2_imshow(output)
```

Thresholding

```python
cv2_imshow(img_gs)
ret,thresh1 = cv2.threshold(img_gs, 125, 255, cv2.THRESH_BINARY)
cv2_imshow(thresh1)
```

Adaptive Thresholding

```python
from matplotlib import pyplot as plt

ret,th1 = cv2.threshold(img_gs,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img_gs,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,2)
th3 = cv2.adaptiveThreshold(img_gs,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)

titles = ['Original Image', 'Global Thresholding (v = 127)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img_gs, th1, th2, th3]

for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
